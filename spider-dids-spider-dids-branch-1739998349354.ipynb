## Welcome to Spider DIDS

Spider DIDS is a database designed to serve as a scientific platform for investigating biological responses to nano-scale coatings (e.g., crystal orientation, defect density, layer count) applied to metal surfaces. Biofilm-DIDS harnesses the power of natural language processing, text mining, data integration, domain ontologies, and an artificial intelligence metadata annotation system. These capabilities empower users to explore integrated datasets derived from experiments as well as existing resources (e.g., materials, transcriptomics, proteomics, metabolomics, methylome, phenotypic, analytic tools, publications, and more).
